```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv("combined.csv")
```

```python
df
```

```python
x= df["Month"]
y1= df["Confirmed Cases"]
y2= df["Deceased Cases"]
y3= df[" Recovered Cases"]
```

```python
plt.figure(figsize=(16,9))
plt.plot(x,y1,color="blue",linewidth=4,marker="o",linestyle="--",label="Confirmed Cases")
plt.plot(x,y2,color="red",linewidth=4,marker="o", linestyle="--",label="Deceased Cases")
plt.plot(x,y3,color="green",linewidth=4,marker="o", linestyle="--",label="Recovered Cases")
plt.title("India's all the Cases", fontsize=25)
plt.xlabel("Month", fontsize=20)
plt.ylabel("Combined Cases", fontsize=20)
plt.legend(loc=2,fontsize=20)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")

width=0.2
plt.figure(figsize=(16,9))
plt.bar(x,y1,color="blue",width=0.6,label="Confirmed Cases")
plt.bar(x,y2,color="red",width=0.6,label="Deceased Cases")
plt.bar(x,y3,color="green",width=0.6,label="Recovered Cases")


plt.grid(color="k",linestyle="-",linewidth="2")
plt.title("India's Confirmed Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Confirmed Cases",fontsize=20)
plt.legend(loc=2,fontsize=20)

plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y1,color="g",marker="*",s=45,alpha=0.8,linewidths=5,edgecolor="blue")
plt.scatter(x,y2,color="g",marker="*",s=45,alpha=0.8,linewidths=5,edgecolor="red")
plt.scatter(x,y3,color="g",marker="*",s=45,alpha=0.8,linewidths=5,edgecolor="green")

plt.title("India's Confirmed Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Confirmed Cases",fontsize=20)
```
