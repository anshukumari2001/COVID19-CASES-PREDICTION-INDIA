```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data= pd.read_csv("state_data.csv")
data
```

```python
x= data["State/UT"]
y1=data["Confirmed"]
y2=data["Active"]
y3=data["Recovered"]
y4=data["Deceased"]
```

```python
plt.figure(figsize=(25,16))

plt.plot(x,y1,color="blue",linewidth=4,marker="o",linestyle="--",label="Confirmed Cases")

plt.title("India's confirmed Cases of all states", fontsize=35)
plt.xlabel("Month", fontsize=30)
plt.xticks(rotation = 90)
plt.ylabel("Confirmed Cases", fontsize=30)
plt.legend(loc=2,fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(25,16))


plt.plot(x,y2,color="red",linewidth=4,marker="o", linestyle="--",label="Active Cases")
plt.title("India's Active Cases of all states", fontsize=35)
plt.xlabel("Month", fontsize=30)
plt.xticks(rotation = 90)
plt.ylabel("Active Cases", fontsize=30)
plt.legend(loc=2,fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(25,16))
plt.plot(x,y3,color="green",linewidth=4,marker="o", linestyle="--",label="Recovered Cases")

plt.title("India's Recovered Cases of all states", fontsize=35)
plt.xlabel("Month", fontsize=30)
plt.xticks(rotation = 90)
plt.ylabel("Recovered Cases", fontsize=30)
plt.legend(loc=2,fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(25,16))
plt.plot(x,y4,color="m",linewidth=4,marker="o", linestyle="--",label="Deceased Cases")
plt.title("India's Deceased Cases of all states", fontsize=35)
plt.xlabel("Month", fontsize=30)
plt.xticks(rotation = 90)
plt.ylabel("Deceased Cases", fontsize=30)
plt.legend(loc=2,fontsize=30)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(25,16))

plt.bar(x, y1, color = "blue",label="Confirmed")
plt.grid(color="k",linestyle="-",linewidth="1")
plt.xlabel("STATES",fontsize=35)
plt.xticks(rotation = 90)
plt.ylabel(" Confirmed CASES",fontsize=35)
plt.title("Covid19 Cases in states of india",fontsize=45)
plt.legend()

plt.show()
```
