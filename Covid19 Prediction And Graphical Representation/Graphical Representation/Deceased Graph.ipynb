```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv("Deceased.csv")
```

```python
df
```

```python
x= df["Month"]
y= df["Deceased Cases"]
```

```python
plt.figure(figsize=(9,4))
axes = plt.gca()
axes.set_ylim([0,160000])
plt.plot(x,y,color="blue")
plt.title("India's Deceased Cases", fontsize=25)
plt.xlabel("Month", fontsize=20)
plt.ylabel("Deceased Cases", fontsize=20)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")

width=0.2
plt.figure(figsize=(9,4))
axes = plt.gca()
axes.set_ylim([0,160000])
plt.bar(x,y,color="r",width=0.6,label="Recovered Cases")


plt.grid(color="k",linestyle="-",linewidth="2")
plt.title("India's Deceased Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Deceased Cases",fontsize=20)
plt.legend(loc=2,fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(9,4))
axes = plt.gca()
axes.set_ylim([0,160000])
plt.scatter(x,y,color="r",marker=".",s=45,alpha=0.8,linewidths=5,edgecolor="r")
plt.title("India's Deceased Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Deceased Cases",fontsize=20)
```

```python
textprops={"fontsize":18}

plt.pie(y,labels=x,radius=4,textprops=textprops,autopct="%0.2f%%" )
plt.show()
```
