```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv("confirmed.csv")
```

```python
df
```

```python
x= df["Month"]
y= df["Confirmed Cases"]
```

```python
plt.figure(figsize=(16,9))
plt.plot(x,y,color="blue")
plt.title("India's Confirmed Cases", fontsize=25)
plt.xlabel("Month", fontsize=20)
plt.ylabel("Confirmed Cases", fontsize=20)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")

width=0.2
plt.figure(figsize=(16,9))
plt.bar(x,y,color="r",width=0.6,label="Confirmed Cases")


plt.grid(color="k",linestyle="-",linewidth="2")
plt.title("India's Confirmed Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Confirmed Cases",fontsize=20)
plt.legend(loc=2,fontsize=20)

plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,color="g",marker="*",s=45,alpha=0.8,linewidths=5,edgecolor="g")
plt.title("India's Confirmed Cases",fontsize=25)
plt.xlabel("Month",fontsize=20)
plt.ylabel("Confirmed Cases",fontsize=20)
```

```python
textprops={"fontsize":18}

plt.pie(y,labels=x,radius=4,textprops=textprops,autopct="%0.2f%%" )
plt.show()
```
