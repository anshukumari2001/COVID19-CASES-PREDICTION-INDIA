```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data= pd.read_csv("state3_data.csv")
data
```

```python
x1= data["State/UT"]
y1= data["Confirmed"]
y2= data["Active"]
y3= data["Recovered"]
y4= data["Deceased"]
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(25,16))

plt.plot(x1,y1,"go--",linewidth=6,markersize=8,label="Confirmed Cases")
plt.plot(x1,y2,"yo-.",linewidth=5,markersize=8,label="Active Cases")
plt.plot(x1,y3,"ko--",linewidth=7,markersize=8,label="Recovered Cases")
plt.plot(x1,y4,"ro--",linewidth=8,markersize=8,label="Deceased Cases")

plt.xlabel("States that are most affected",fontsize=30)
plt.ylabel("Cases",fontsize=30)
plt.title("Covid19 Impact ",fontsize=45)
plt.grid(color="k",linestyle="-",linewidth=1)
plt.legend(loc=4,fontsize=25)
plt.show()
```
