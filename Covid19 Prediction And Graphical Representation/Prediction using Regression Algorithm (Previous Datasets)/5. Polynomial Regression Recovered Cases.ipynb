```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data= pd.read_csv("Prediction.csv")
data
```

```python
real_x= data.iloc[:,1].values
real_y= data.iloc[:,3].values
real_x= real_x.reshape(-1,1)
real_y= real_y.reshape(-1,1)
```

```python
training_x,testing_x,training_y,testing_y = train_test_split(real_x,real_y,test_size=0.2,random_state=0) 
```

```python
poly_reg= PolynomialFeatures(degree=2)
real_x_reg= poly_reg.fit_transform(training_x)
poly_reg.fit(real_x_reg,training_y)
Lin2= LinearRegression()
Lin2.fit(real_x_reg,training_y)
```

```python
pred_y=Lin2.predict(poly_reg.fit_transform(testing_x))
```

```python
Lin2.predict(poly_reg.fit_transform([[81]]))
```

```python
plt.scatter(training_x,training_y,color="red")
plt.plot(training_x,Lin2.predict(poly_reg.fit_transform(training_x)),color="blue")
axes=plt.gca()
axes.set_ylim([0,160000])
plt.title("India's Recovered Cases")
plt.xlabel("Date from 1 april")
plt.ylabel("Number of Recovered cases")
plt.show()
```

```python
plt.scatter(testing_x,testing_y,color="red")
plt.plot(training_x,Lin2.predict(poly_reg.fit_transform(training_x)),color="blue")
axes=plt.gca()
axes.set_ylim([0,160000])
plt.title("India's Recovered Cases")
plt.xlabel("Date from 1 april")
plt.ylabel("Number of Recovered cases")
plt.show()
```

```python
r2_score(testing_y,pred_y)*100 #Check Accuracy
```
