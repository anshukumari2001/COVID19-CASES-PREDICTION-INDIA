```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Confirmed_statewise.csv")
data
```

```python
x=data["Recent_dates"]
y1=data["Maharashtra"]
y2=data["Gujrat"]
y3=data["Delhi"]
y4=data["Tamil Nadu"]
y5=data["Rajasthan"]
y6=data["Madhya_pradesh"]
y7=data["Uttar_pradesh"]
y8=data["Andhra_pradesh"]
y9=data["West_bengal"]
y10=data["Punjab"]
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.plot(x,y1,color="r", marker="o", linestyle="--",linewidth=3,markersize=8,label="Maharashtra")
plt.plot(x,y2,color="green", marker="o", linestyle="--",linewidth=3,markersize=8,label="Gujrat")
plt.plot(x,y3,color="blue", marker="o", linestyle="--",linewidth=3,markersize=8,label="Delhi")
plt.plot(x,y4,color="m", marker="o", linestyle="--",linewidth=3,markersize=8,label="Tamil Nadu")
plt.plot(x,y5,color="black", marker="o", linestyle="--",linewidth=3,markersize=8,label="Rajasthan")
plt.plot(x,y6,color="yellow", marker="o", linestyle="--",linewidth=3,markersize=8,label="Madhya_pradesh")
plt.plot(x,y7,color="indigo", marker="o", linestyle="--",linewidth=3,markersize=8,label="Uttar_pradesh")
plt.plot(x,y8,color="pink", marker="o", linestyle="--",linewidth=3,markersize=8,label="Andhra_pradesh")
plt.plot(x,y9,color="grey", marker="o", linestyle="--",linewidth=3,markersize=8,label="West_bengal")
plt.plot(x,y10,color="brown", marker="o", linestyle="--",linewidth=3,markersize=8,label="Punjab")

plt.xlabel("Month",fontsize=20)
plt.xticks(rotation=70)
plt.ylabel("Confirmed Cases",fontsize=20)
plt.title("India's most affected states",fontsize=25)
plt.grid(color="k",linestyle="-",linewidth=1)
plt.legend(loc=2)
plt.show()
```
